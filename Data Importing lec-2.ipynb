```python
import numpy as np
import pandas as pd
import pickle
```

```python
def load_csv(filepath):
  data=[]
  col=[]
  checkcol=False
  with open(filepath) as f:
    for val in f.readlines():
      val=val.replace("\n","")
      val=val.split(",")
      if checkcol is False:
        col=val
        checkcol=True
      else:
        data.append(val)
  df=pd.DataFrame(data,columns=col)
  return df
```

```python
myData=load_csv("/content/100 sales.csv")#importing dataset
myData
```
