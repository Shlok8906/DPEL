Multiindex and index Hierachy



```python
#index levels
outside=['G1','G1','G1','G2','G2','G2']
inside=[1,2,3,1,2,3]
hier_index=list(zip(outside,inside))
hier_index=pd.MultiIndex.from_tuples(hier_index)
```

```python
hier_index
```

```python
df=pd.DataFrame(np.random.randn(6,2),index=hier_index,columns=['A','B'])
df
```

```python
df.loc['G1']
```

```python
df.loc['G1'].loc[1]
```

```python
df.index.names#to get the name of the index
```

```python
df.index.names=['Group','Num']
df
```

```python
df.xs('G1')#to get specific group
```

```python
df.xs(('G1',1))#gives info of 1 from G1

```

```python
df.xs(1,level='Num')#to get info
```

```python
data1={
    "name":["Sally","Marry","Jhon"],
    "Age":[50,40,30]
}
data2={
    "name":["Sally","peter","Mickey"],
    "Age":[77,44,22]
}
df1=pd.DataFrame(data1)
df2=pd.DataFrame(data2)
df1
#make a dataset maunally
```

```python
df2
```

```python
newdf=df1.merge(df2,how='right')
newdf
```

```python
newdf=df1.merge(df2,how='cross')
newdf
```

```python
left=pd.DataFrame({'Key':['K0',"K1","K2","K3"],
                   'A':['A0','A1',"A2","A3"],
                   'B':['B0','B1',"B2","B3"]
                   })
right=pd.DataFrame({'Key':['K0',"K1","K2","K3"],
                   'C':['C0','C1',"C2","C3"],
                   'D':['D0','D1',"D2","D3"]
                   })
```

```python
pd.merge(left,right,how='inner',on='Key')
#meges on the comman. indexx
```

```python
pd.merge(left,right,how='outer',on='Key')
```
